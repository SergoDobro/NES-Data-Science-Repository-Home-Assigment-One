## 3. Средняя сладость клубники.

Ваша бабушка решила поспорить с соседкой, у кого клубника более сладкая. Они посадили кусты и потом оценили сладость клубники, собранной с каждого из кустов, по шкале от 0 до 10. Ваша задача - посчитать среднее для двух массивов и определить, у кого слаще клубника в среднем.

```python
grandma_strawberries = [5, 7, 10, 9, 3, 10]
neighbour_strawberries = [10, 6, 2, 3]

def count_sugarness(array_of_strawberries):
  return sum(array_of_strawberries)/len(array_of_strawberries)

grandma_strawberries_suggarness = count_sugarness(grandma_strawberries);
neighbour_strawberries_suggarness = count_sugarness(neighbour_strawberries);

print(f'grandma strawberries  suggarness: {grandma_strawberries_suggarness}')
print(f'neighbour strawberries  suggarness: {neighbour_strawberries_suggarness}')

if grandma_strawberries_suggarness > neighbour_strawberries_suggarness:
  print('grandma\'s strawberries on average are more sugary!')
elif grandma_strawberries_suggarness < neighbour_strawberries_suggarness:
  print('neighbour\'s strawberries on average are more sugary!')
else:
  print('grandma\'s and neighbour\'s strawberries on average are equally sugary!')
```

1. Самое длинное слово в строке
2. Количество слов в строке
3. Переворачивает каждое слово в строке задом наперёд
4. Удаляет все цифры из строки
5. Заменяет все буквы верхнего регистра на нижний, а нижнего регистра на верхний
6. Проверяет, является ли строка палиндромом (без учета пробелов, регистра и пунктуации, те только буквы и цифры)
7. Если строка не является палиндромом, то выводит первые 5 символов строки в обратном порядке

Т.е.  первая строка - самое длинное слово в исходной строке, вторая строка - количество слов в исходной строке и.т.д.

```python
def get_extended_string_info(input_string: str):
  words: list[str] = input_string.split()#list of words

  #print(max([len(x) for x in input_string.split()])) this is strange

  longest_word: str = max(words, key=len) #longest word

  words_count = len(words)

  reversed_words: list = [w[::-1] for w in words]
  reversed_str = ' '.join(reversed_words)

  numberless_input_string = ''.join([x for x in input_string if x.isdigit() == False])

  casereversed_input_string = input_string.swapcase()

  _input_string_only_letters_or_digits = ''.join(letter for letter in input_string if letter.isalnum() or letter.isdigit())
  isPolyndrom_input_string = _input_string_only_letters_or_digits.lower()  == _input_string_only_letters_or_digits[::-1].lower()
  first_symbols_reversed = ''
  if not isPolyndrom_input_string:
    first_symbols_reversed = (input_string[4:0:-1] + input_string[0])


  return longest_word, words_count, reversed_str, numberless_input_string, casereversed_input_string, isPolyndrom_input_string, first_symbols_reversed;
```

```python

print(get_extended_string_info("Testing tester to 12 42 and 42F"))
print(get_extended_string_info("L5o5L"))
print(get_extended_string_info("Hello,o oLLeh"))
```
